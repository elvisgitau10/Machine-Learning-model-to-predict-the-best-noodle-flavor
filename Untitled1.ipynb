```python
variety = df['Variety']
```

```python
variety
```

```python
columns = df.columns.str.lower()
```

```python
# Combine variety into a single text
```

```python
variety_text = " ".join(variety)
```

```python
variety_text
```

```python
#Convert everything into lower text
```

```python
variety_tetx_lower = variety_text.lower()
```

```python
variety_tetx_lower
```

```python
# We need to remove punctuations
# Lets create a list of punctuations
```

```python
punc_list = list(string.punctuation)
punc_list
```

```python
# Turn everything in variety_text_lower into a list of single letters.
```

```python
variety_text_clean ="".join([ char for char in variety_tetx_lower if char not in punc_list])
```

```python
variety_text_clean
```

```python
# We use nltk to tokenize the variety_text_clean.
# Tokenization is the process of breaking the given text, into smaller units called tokens.
# In other words "word segmentation"
```

```python
nltk.download('punkt')
variety_tokens = nltk.tokenize.word_tokenize(variety_text_clean)
variety_tokens
```

```python
# Lets look at the frequency of the tokens
```

```python
variety_counter = collections.Counter(variety_tokens)
variety_counter
```

```python
# To get an arrangement
```

```python
variety_counter.most_common()
```

```python
# Take note of the spelling variation in the text.
# For instance, flavour and flavor
```

```python
is_spicy = []
hot_flavor = ["spicy",
              "Curry",
             "Hot",
             "tom",
             "chili",
             "chilli",
             "kari",
             "pedas",
             "laksa"] 
for variety in df['Variety'].str.lower():
    if any (flavor in variety for flavor in hot_flavor):
         is_spicy.append(1)
    else:
        is_spicy.append(0)
df["isSpicy"] =is_spicy
df
```

```python
has_chicken =[]

for variety in df['Variety'].str.lower():
    if ('chicken' in variety) or ("ayam" in variety):
        has_chicken.append(1)
    else:
        has_chicken.append(0)
df["hasChicken"] = has_chicken
df
    
```

```python
df.sample(10)
```

```python
has_beef =[]

for variety in df['Variety'].str.lower():
    if ('beef' in variety) or ("cow" in variety):
        has_beef.append(1)
    else:
        has_beef.append(0)
df["hasBeef"] = has_beef
df
```

```python
df.sample(10)
```

```python
has_seafood = []

seafood_flavor = ["seafood",
                 "fish",
                  "shrimp",
                 "crab",
                 "prawn"]
for variety in df['Variety'].str.lower():
    if any (flavor in variety for flavor in seafood_flavor):
         has_seafood.append(1)
    else:
        has_seafood.append(0)
df["hasSeafood"] = has_seafood
df
```

```python
df.sample(10)
```
