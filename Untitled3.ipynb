```python
# import libraries
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
# import metrics
from sklearn.metrics import confusion_matrix, classification_report
```

```python
df = pd.read_csv("The-Big_List_clean_extended_final.csv")
df
```

```python
# Take a look at Stars column
df["Stars"].value_counts()
```

```python
#import ploting library
import matplotlib.pyplot as plt
plt.hist(df["Stars"])
```

```python
#We need to categorize "Stars" into categories.
df["Stars"].median()
```

```python
len(df)
```

```python
len(df[df["Stars"]> 3.75])
```

```python
df["bin_stars"]=pd.qcut(df["Stars"],q=2, labels =[0,1])
```

```python
df
```

```python
df["bin_stars"].value_counts()
```

```python
# prepare independent and dependent variables
```

```python
x = df.drop(["bin_stars", "Stars"], axis = 1)
y = df["bin_stars"]
```

```python
# split data set into train and test
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,
                                                   test_size = 0.2, 
                                                    stratify=y)
```

```python
# Lets set a baseline with a DummyClassifier
```

```python
dummy = DummyClassifier()
dummy.fit(x_train, y_train)
dummy_pred = dummy.predict(x_test)
```

```python
# Assess the performance of DummyClassifier
confusion_matrix(y_test, dummy_pred)
```

```python
print(classification_report(y_test, dummy_pred))
```

```python
# start with logisticRegression
```

```python
logr = LogisticRegression()
logr.fit(x_train, y_train)
logr_pred = logr.predict(x_test)
```

```python
confusion_matrix(y_test, logr_pred)
```

```python
print(classification_report(y_test, logr_pred))
```

```python
#DecisionTreeClassifier
```

```python
tree = DecisionTreeClassifier()
tree.fit(x_train, y_train)
tree_pred = tree.predict(x_test)
```

```python
confusion_matrix(y_test, tree_pred)
```

```python
print(classification_report(y_test,tree_pred))
```

```python
# use RandomForestClassifier
```

```python
forest = RandomForestClassifier()
forest.fit(x_train, y_train)
forest_pred = forest.predict(x_test)
```

```python
confusion_matrix(y_test, forest_pred)
```

```python
print(classification_report(y_test,forest_pred))
```
