```python
# import libraries
```

```python
import pandas as pd
```

```python
df = pd.read_csv("The-Big_List_clean_extended.csv")
```

```python
df
```

```python
# get value_counts of Brand 
```

```python
df["Brand"].value_counts()
```

```python
# preview the first 20 brands
```

```python
top_thirty_brands = df["Brand"].value_counts()[:30].index
```

```python
top_thirty_brands
```

```python
# As you go down the group the frequency of the brands decreases.
# We convert the rest of the brands to others.
# The objective here is to reduce the number of unique values.
```

```python
# create a new column called newBrand
```

```python
new_brand = []
for brand in df["Brand"]:
    if brand in top_thirty_brands:
        new_brand.append(brand)
    else :
        new_brand.append("other")
df["newBrand"] = new_brand
df
```

```python
df["newBrand"].value_counts()
```

```python
# examine style
```

```python
top_four_styles = df["Style"].value_counts().index[:4]
```

```python
# loop through style and create new style
```

```python
new_style = []
for style in df["Style"]:
    if style in top_four_styles:
        new_style.append(style)
    else :
        new_style.append("other")
df["newstyle"] = new_style
df
```

```python
df["newstyle"].value_counts()
```

```python
top_ten_countries = df["Country"].value_counts()[:10].index
```

```python
top_ten_countries
```

```python
new_country = []
for country in df["Country"]:
    if country in top_ten_countries:
        new_country.append(country)
    else :
        new_country.append("other")
df["newcountry"] = new_country
df
```

```python
# We will now dummyfy our data
```

```python
brand_dummy = pd.get_dummies(df["newBrand"], drop_first = True,
                            prefix ="from")
brand_dummy
```

```python
style_dummy = pd.get_dummies(df["newstyle"], drop_first = True,
                            prefix ="is")
style_dummy
```

```python
country_dummy = pd.get_dummies(df["newcountry"], drop_first = True,
                            prefix ="in")
country_dummy
```

```python
df
```

```python
# We need to drop Brand, variety, style and country.
```

```python
df.columns
```

```python
df_temp = df.drop(['Brand', 'Variety', 'Style', 'Country','newBrand', 'newstyle',
       'newcountry'], axis =1)
df_temp
```

```python
# We need to combine this with the rest of the dummies we once obtained.
```

```python
df_final = pd.concat([df_temp,brand_dummy,style_dummy,country_dummy],
                    axis =1 )
```

```python
df_final
```

```python
df_final.to_csv("The-Big_List_clean_extended_final.csv", index= None)
```

###  Summary

1. Changed low frequency values into "others"
2. Dummified the new columns.
3. Dropped unnecessary columns.
4. Joined everything into a final DataFrame.
