```python
# Import libraries
import pandas as pd
```

```python
df = pd.read_excel("The-Big-List-20210117.xlsx")
```

```python
df
```

```python
#figure out rows with missing data
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df = df.drop('T',axis=1)
```

```python
# We have droped 'T' because it does not contain any useful information.
#Further review number is not important
```

```python
df = df.drop(['Review #'], axis=1)
```

```python
df
```

```python
df.dtypes
```

```python
#Lets take a look at the stars column
```

```python
df['Stars']
```

```python
# Use value_counts method to examine unique values
```

```python
df['Stars'].value_counts()
```

```python
# consider replacing the fractions into proper float
```

```python
# 4.5/5 ->4.5
# 5/2.5 -> 2.5
# 3.5/2.5 -> 2.5
# 4.25/5 -> 5
```

```python
df["Stars"]= df["Stars"].replace("4.5/5",4.5)
df["Stars"]= df["Stars"].replace("5/2.5",2.5)
df["Stars"]= df["Stars"].replace("3.5/2.5",2.5)
df["Stars"]= df["Stars"].replace("4.25/5",5)

df["Stars"].value_counts()
```

```python
# What remains are strings and date time
```

```python
# clean "stars" column
# One way is by converting anything that is not a number to na
```

```python
df["Stars"]= pd.to_numeric(df["Stars"], errors= "coerce")
```

```python
df["Stars"].value_counts()
```

```python
#Check NaN in rows
```

```python
df.isnull().sum()
```

```python
#Remove rows with NaN
df = df.dropna()
df.isnull().sum()
```

```python
# Take a look at df
```

```python
df
```

```python
df.to_csv("The-Big-List-clean.csv", index = None)
```
